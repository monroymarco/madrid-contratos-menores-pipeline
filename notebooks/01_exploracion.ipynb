```python
import pandas as pd
df = pd.read_csv('../data/raw/contratos_menores_2026_csv.csv', sep=';', encoding='latin-1')
df.head()
df.info()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.upper()
    .str.replace(' - ', '_', regex=False)    
    .str.replace(' ', '_')
    .str.replace('.', '', regex=False)
)
df
```

```python
cols_importe = ['IMPORTE_LICITACION_IVA_INC', 'IMPORTE_ADJUDICACION_IVA_INC']  # columnas de dinero

for col in cols_importe:                                                        # recorre cada columna de importe
    df[col] = (                                                                 # sobrescribe la columna ya limpia
        df[col].astype(str)                                                     # asegura texto
        .str.replace(r'[^\d,.-]', '', regex=True)                               # elimina símbolos raros como € o 
        .str.replace('.', '', regex=False)                                      # quita separador de miles
        .str.replace(',', '.', regex=False)                                     # convierte coma decimal a punto
        .pipe(pd.to_numeric, errors='coerce')                                   # convierte a número
    )
```
