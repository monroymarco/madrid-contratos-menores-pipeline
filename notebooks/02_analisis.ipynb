```python
import pandas as pd
df = pd.read_csv('contratos_menores_clean.csv')
```

```python
df.head()
```

```python
kpi_gasto_mes = df.groupby('MES', as_index=False)['IMPORTE_ADJUDICACION_IVA_INC'].sum().sort_values('MES')
```

```python
kpi_gasto_mes
```

```python
df['FECHA_DE_ADJUDICACION'].max()
```

```python
df['FECHA_DE_ADJUDICACION'].min()
```

```python
kpi_contratos_mes = df.groupby('MES', as_index=False)['N_DE_REGISTRO_DE_CONTRATO'].count().sort_values('MES')
```

```python
kpi_contratos_mes
```

```python
kpi_ticket_promedio = df.groupby('MES', as_index=False)['IMPORTE_ADJUDICACION_IVA_INC'].mean().sort_values('MES')
```

```python
kpi_ticket_promedio
```

```python
kpi_check = kpi_gasto_mes['IMPORTE_ADJUDICACION_IVA_INC'] / kpi_contratos_mes['N_DE_REGISTRO_DE_CONTRATO']
kpi_check
```

```python
kpi_top_empresas_contratos = (df.groupby('RAZON_SOCIAL_ADJUDICATARIO')
    ['N_DE_REGISTRO_DE_CONTRATO']
    .count()
    .sort_values(ascending=False)
    .head(10)
)                       
kpi_top_empresas_contratos
```

```python
kpi_top_empresas_monto = (df.groupby('RAZON_SOCIAL_ADJUDICATARIO')['IMPORTE_ADJUDICACION_IVA_INC']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
kpi_top_empresas_monto
```

```python
kpi_mediana_mes = (df.groupby('MES', as_index=False)['IMPORTE_ADJUDICACION_IVA_INC']
    .median()
    .sort_values('MES')
)
kpi_mediana_mes
```

```python
kpi_top_organo_contratacion_mes = (df.groupby(['MES', 'ORGANO_DE_CONTRATACION'])['IMPORTE_ADJUDICACION_IVA_INC']
    .sum()
    .reset_index()
    .sort_values(['MES', 'IMPORTE_ADJUDICACION_IVA_INC'], ascending=[True, False])
    .groupby('MES')
    .head(5)
)
kpi_top_organo_contratacion_mes
```

```python
kpi_tipo_contrato_mes = (df.groupby(['MES', 'TIPO_DE_CONTRATO'])['IMPORTE_ADJUDICACION_IVA_INC']
    .sum()
    .reset_index()
    .sort_values(['MES','IMPORTE_ADJUDICACION_IVA_INC'], ascending=[True, False])
)
kpi_tipo_contrato_mes
```

```python
kpis = {
    'kpi_gasto_mes': kpi_gasto_mes,
    'kpi_contratos_mes': kpi_contratos_mes,
    'kpi_ticket_promedio': kpi_ticket_promedio,
    'kpi_top_empresas_contratos': kpi_top_empresas_contratos,
    'kpi_top_empresas_monto': kpi_top_empresas_monto,
    'kpi_mediana_mes': kpi_mediana_mes,
    'kpi_top_organo_contratacion_mes': kpi_top_organo_contratacion_mes,
    'kpi_tipo_contrato_mes': kpi_tipo_contrato_mes
}
for name, df_kpi in kpis.items():
    df_kpi.reset_index().to_csv(
        f'../output/kpis/{name}.csv',
        index=False
    )
```
